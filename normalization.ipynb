```python
sample2_refandsample= pd.read_csv('measurments again/abs for ref and sample again.csv')
sample1_refandsample= pd.read_csv('ABS results for TPA measurments both ref and sample .csv')
sample34_refandsample= pd.read_csv('TPA for sample 3 and 4/abs for big and very small nps with reference sample.csv')

sample1_refandsample
```

```python
plt.plot(sample2_refandsample.iloc[:,0],sample2_refandsample.iloc[:,1])
plt.plot(sample1_refandsample.iloc[:,8], sample1_refandsample.iloc[:,9])
plt.plot(sample34_refandsample.iloc[:,0], sample34_refandsample.iloc[:,1])
plt.plot(sample34_refandsample.iloc[:,4], sample34_refandsample.iloc[:,5])

plt.xlim(300,700)
```

```python
plt.plot(sample2_refandsample.iloc[:,2],sample2_refandsample.iloc[:,3])
plt.plot(sample1_refandsample.iloc[:,6], sample1_refandsample.iloc[:,7])
plt.plot(sample34_refandsample.iloc[:,2], sample34_refandsample.iloc[:,3])
plt.xlim(300,700)
plt.ylim(0,0.7)
```

```python
def min_max_normalize(spectrum):
    min_val = np.min(spectrum)
    max_val = np.max(spectrum)
    return (spectrum - min_val) / (max_val - min_val)

sample2_absorption_normalized = min_max_normalize(sample2_refandsample.iloc[:, 1])
sample1_absorption_normalized = min_max_normalize(sample1_refandsample.iloc[:, 9])
sample34_absorption1_normalized = min_max_normalize(sample34_refandsample.iloc[:, 1])
sample34_absorption2_normalized = min_max_normalize(sample34_refandsample.iloc[:, 5])

# Plot the normalized spectra
# plt.figure(figsize=(10, 6))

plt.plot(sample2_refandsample.iloc[:, 0], sample2_absorption_normalized, label='Sample 2')
plt.plot(sample1_refandsample.iloc[:, 8], sample1_absorption_normalized, label='Sample 1')
plt.plot(sample34_refandsample.iloc[:, 0], sample34_absorption1_normalized, label='Sample 34 - 1')
plt.plot(sample34_refandsample.iloc[:, 4], sample34_absorption2_normalized, label='Sample 34 - 2')

plt.xlim(300,700)
plt.xlabel('Wavelength')
plt.ylabel('Normalized Absorption')
plt.title('Normalized Absorption Spectra')
plt.legend()
plt.show()
```
